```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df['overweight'] = df['weight']/((df['height']/100)*(df['height']/100))
df['overweight'] = pd.cut(df['overweight'],bins = [0,24,300], labels=[0,1])
```

```python
df
```

```python
df['cholesterol'] = pd.cut(df['cholesterol'],bins = [0,1,3], labels=[0,1])
df
```

```python
df['gluc'] = pd.cut(df['gluc'],bins = [0,1,3], labels=[0,1])
df
```

```python
d = pd.melt(df, value_vars=['cholesterol', 'gluc', 'smoke', 'alco','active', 'overweight'])
d
```
